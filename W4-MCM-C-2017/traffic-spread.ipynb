```python
# calculate distance based on position
def cal_distance(x1, y1, x2, y2):
    dis = ((x1 - x2)**2 + (y1 - y2)**2)**0.5
    if (dis < 20):
        return 20
    
    return dis

def cal_distance2(x1, y1, x2, y2):
    return ((x1 - x2)**2 + (y1 - y2)**2)**0.5
```

```python
# Define hypothesis and loss function

def hypo(cities, milepost):
    predict = 0.
    for city in cities:
        predict += city.impact * city.pop / cal_distance(city.x, city.y, milepost.x, milepost.y)
        
    return predict


def loss_train(cities, mileposts):
    total_cost = 0.
    for i, milepost in enumerate(mileposts):
        if milepost.volume != 0:
            total_cost += (hypo(cities, milepost) - milepost.volume)**2
        
    return total_cost / 2 / milepost_num

def loss(cities, milepost):
    total_cost = 0.
    for i, milepost in enumerate(mileposts):
        total_cost += (hypo(cities, milepost) - milepost.volume)**2
        
    return total_cost / 2 / milepost_num
```

```python
# Try to find a good initialization
I = np.random.random((1000, 58)) / 7
min_loss = 10**6
min_index = 0
for i in range(1000):
    for j in range(len(cities)):
        cities[j].impact = I[i, j]
    
    cur_loss = loss_train(cities, mileposts)
    if cur_loss < min_loss:
        min_loss = cur_loss
        min_index = i
    
print(min_loss)
print(min_index)
print("big loss: %ld" % (loss_train(cities, mileposts)))
```

```python
# test the initialization
for j in range(len(cities)):
    cities[j].impact = I[min_index, j]

avg_accu = 0.
count = 0
for i in range(len(mileposts)):
    if mileposts[i].volume == volume2[i]:
        continue
        
    avg_accu += abs(hypo(cities, mileposts[i]) - volume2[i]) / volume2[i]
    count += 1
        
avg_accu = avg_accu / count
print("Test: %d nodes, average accuracy: %lf" % (count, 1 - avg_accu))
```

```python
# Gradient Descent Optimization
for city in cities:
    city.impact = random.random() / 10

epochs_num = 70
learning_rate = 0.0000002

print("At start, loss: %ld" % (loss_train(cities, mileposts)))
for index in range(epochs_num):
    for i in range(len(cities)):
        grad = 0.
        for milepost in mileposts:
            if milepost.volume == 0:
                continue
            a = (hypo(cities, milepost) - milepost.volume) / cal_distance(cities[i].x, cities[i].y, milepost.x, milepost.y) * cities[i].pop
            grad = grad + a
            
        cities[i].impact = cities[i].impact - learning_rate * grad / milepost_num
    
    if index % 10 == 0:
        print("Epochos: %d, loss: %ld" % (index, loss_train(cities, mileposts)))
```
